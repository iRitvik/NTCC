```python
import pandas as pd
```

```python
data = pd.read_excel("Unique Reviews.xlsx")
```

```python
data.info()
```

```python
data["Review"] = data["Review"].str.lower()
```

```python
import string
data["Review"] = data["Review"].str.replace(string.punctuation, "")
```

```python
data
```

```python
data.columns
```

```python
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
```

```python
stop_words = stopwords.words("english")
data['Review'] = data["Review"].astype(str)  # Convert the review
data["Review"] = data["Review"].apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]))
```

```python
data
```

```python
data["review_tokens"] = data["Review"].apply(lambda x: x.split())
```

```python
data
```

```python
import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import streamlit as st
```

```python


# Step 1: Score the Reviews using ASBA
sid = SentimentIntensityAnalyzer()

def get_sentiment_score(review):
    scores = sid.polarity_scores(review)
    return scores['compound']

data['Sentiment_Score'] = data['Review'].apply(get_sentiment_score)
```

```python
data.to_excel("final_scores.xlsx")
```
